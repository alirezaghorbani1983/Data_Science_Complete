```python
import requests as req
from bs4 import BeautifulSoup

page = req.get("https://en.wikipedia.org/wiki/Car")
if page.status_code == 200:
    print("page exists")
    print(page.headers)
else:
    print("page doesn't exist")
```

```python
# use metode text show structure
import requests as req
from bs4 import BeautifulSoup

page = req.get("https://en.wikipedia.org/wiki/Car")
if page.status_code == 200:
    print("page exists")
    print(page.text)
else:
    print("page doesn't exist")
```

```python
# use metode content show structure
import requests as req
from bs4 import BeautifulSoup

page = req.get("https://en.wikipedia.org/wiki/Car")
if page.status_code == 200:
    print("page exists")
    print(page.content)
else:
    print("page doesn't exist")
```
